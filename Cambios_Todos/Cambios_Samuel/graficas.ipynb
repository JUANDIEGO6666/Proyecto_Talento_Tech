```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

ruta= Path('Cambios_Todos')/'archivo.csv'
demog = pd.read_csv('/home/cmdml/Documentos/Programción/Ejercicio-TalentoTech/Proyecto/Proyecto_Talento_Tech/Cambios_Todos/Cambios_Andrés/archivo_corregido.csv')
len(demog)
```

```python
plt.figure(figsize = (8,5))
plt.plot(demog["Año"], demog["Población Total"], color = "magenta")
plt.plot(demog["Año"], demog["Población por área Urbana"], color = "skyblue")
plt.plot(demog["Año"], demog["Población Centros Poblados y Rural Disperso"], color = "orange")
plt.title("Crecimiento Poblacional Progresivo")
plt.xlabel("Año")
plt.ylabel("Personas")
plt.legend(["Población Total",
            "Población Urbana",
            "Población Rural"])

plt.savefig("Crecimiento.png")
plt.show()
```

```python
plt.figure(figsize = (8,5))
plt.plot(demog["Año"], demog["Relación hombre-mujer"], color = "magenta")
plt.plot(demog["Año"], demog["Índice dedependencia"], color = "skyblue")
plt.plot(demog["Año"], demog["Índice deenvejecimiento"], color = "#6b6231")
plt.plot(demog["Año"], demog["Índice deinfancia"], color = "#00ffd5")
plt.plot(demog["Año"], demog["Índice dejuventud"], color = "orange")
plt.plot(demog["Año"], demog["Índice de vejez"], color = "#6f7919")


plt.title("Variación de los índices a través del tiempo")
plt.xlabel("Año")
plt.legend(["Relación Hombre-Mujer",
            "Índ. de Dependencia",
            "Índ. de Envejecimiento",
            "Índ. de Infancia",
            "Índ. de Juventud",
            "Índ. de Vejez"])

plt.savefig("IndicesVarios.png")
plt.show()
```
